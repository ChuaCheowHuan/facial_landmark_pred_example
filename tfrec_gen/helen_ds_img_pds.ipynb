##Read images

```python
def read_img(file):
    """ 
    Read a single image file
    """
    split_line = file.split("/")
    split_line = split_line[-1].replace(".jpg", "")
    img_name = split_line

    with open(file, 'rb') as img_file:
        img = img_file.read()

    return img_name, img    
```

```python
def read_imgs(folder_path):
    """ 
    Read image files from a single folder.
    """
    files = os.listdir(folder_path)
    imgs_name = []
    imgs = []
    for f in files:
        img_name, img = read_img(folder_path + f)
        imgs_name.append(img_name)
        imgs.append(img)                
    
    return imgs_name, imgs
```

```python
def read_imgs_folder(folder_list, folder_num):
    """
    Read image files from multiple folders.
    """
    imgs_name_list = []
    imgs_list = []
    for i, data_sub_folders in enumerate(folder_list):
        if i>0:
            folder_path = data_folder + data_sub_folders + "/" + data_sub_folders + "/"
            imgs_name, imgs = read_imgs(folder_path)
            imgs_name_list.append(imgs_name)
            imgs_list.append(imgs) 

            # Write to pandas dataframe.
            names_df = pd.DataFrame(imgs_name)
            names_df.columns = ['file_name']    
            imgs_df = pd.DataFrame(imgs)
            imgs_df.columns = ['img']

            # Concat Names & images df.
            names_p_imgs_df = pd.concat([names_df, imgs_df], axis=1)     
            names_p_imgs_df.to_pickle(path + "helen_pickle/" + "helen_ds_img_" + str(i) + ".pickle")
            print(names_p_imgs_df.head())

    return imgs_name_list, imgs_list    
```
